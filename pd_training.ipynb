merge() - Column match

```python
izq = pd.DataFrame({
    'key': ['k0', 'k1', 'k2', 'k3'],
    'A': ["A0", "A1", "A2", "A3"],
    "B": ["B0", "B1", "B2", "B3"]
})

izq
```

```python
der = pd.DataFrame({
    'key': ['k0', 'k1', 'k2', 'k3'],
    'C': ["C0", "C1", "C2", "C3"],
    "D": ["D0", "D1", "D2", "D3"]
})

der
```

```python
izq.merge(der, on='key')
```

```python
izq = pd.DataFrame({
    'key_1': ['k0', 'k1', 'k2', 'k3'],
    'A': ["A0", "A1", "A2", "A3"],
    "B": ["B0", "B1", "B2", "B3"]
})

der = pd.DataFrame({
    'key_2': ['k0', 'k1', 'k2', 'k3'],
    'C': ["C0", "C1", "C2", "C3"],
    "D": ["D0", "D1", "D2", "D3"]
})
```

```python
izq.merge(der, left_on='key_1', right_on='key_2')
```

```python
der = pd.DataFrame({
    'key_2': ['k0', 'k1', 'k2', np.nan],
    'C': ["C0", "C1", "C2", "C3"],
    "D": ["D0", "D1", "D2", "D3"]
})
```

```python
izq.merge(der, left_on='key_1', right_on='key_2')
```

The 'how=' argument can take 4 values: inner (by default), outer, left and right.

```python
izq.merge(der, left_on='key_1', right_on='key_2', how='left')
```

```python
izq.merge(der, left_on='key_1', right_on='key_2', how='right')
```

join() - Index match

```python
izq = pd.DataFrame({
    'A': ["A0", "A1", "A2"],
    "B": ["B0", "B1", "B2"]
}, index=['k0', 'k1', 'k2'])

der = pd.DataFrame({
    'C': ["C0", "C1", "C2"],
    "D": ["D0", "D1", "D2"]
}, index=['k0', 'k2', 'k3'])
```

```python
izq.join(der, how='inner')
```

```python
izq.join(der, how='outer')
```

```python
izq.join(der, how='left')
```

```python
izq.join(der, how='right')
```
